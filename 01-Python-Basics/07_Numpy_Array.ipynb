## 1. Introduction

NumPy (Numerical Python) is one of the most important libraries in Python for scientific computing, data analysis, and AI/ML. It provides a powerful array-based data structure called the NumPy array, which is essential for handling and performing fast mathematical operations on large datasets.

What is an Array?
An array is a data structure that stores a collection of items, usually numbers, where each item is accessed by its position (or index) in the sequence. Unlike lists in Python, arrays require all elements to be of the same data type (such as all integers or all floats).
