# Introduction to Python Variables and Data Types

**Variables** in Python are like labels you attach to values so you can use, manipulate, and refer to those values later in your code.

**Data types** tell Python what kind of value a variable holds (like a number, text, or a collection of values). Knowing data types helps you store information correctly and use all the right operations for each type.

---


## Common Python Data Types

- **int**: Whole numbers (e.g., 5, -12)
- **float**: Decimal numbers (e.g., 3.14, -0.5)
- **str**: Text data called strings (e.g., "Hello", 'World')
- **bool**: True or False values (e.g., True, False)
- **list**: Ordered, changeable collection (e.g., [1, 2, 3])
- **tuple**: Ordered, unchangeable collection (e.g., (4, 5, 6))
- **set**: Unordered, unique items collection (e.g., {'a', 'b'})
- **dict**: Key-value pairs (e.g., {'name': 'Alice', 'age': 23})

---


```python
# int and float variables
int_var = 10
float_var = 3.1415

# Addition
result = int_var + float_var  # 13.1415
print(result)
```

---
## Summary & Tips

- Choose the right data type for your task: use lists and dictionaries for group data, numbers for calculations, strings for texts.
- Think about whether you need items to change (choose lists and dicts) or stay fixed (choose tuples).
- Avoid mixing data types in one variable unless you mean to!
- Print your variables to check their values as you work.
- Beginners often forget:
  - To use the correct bracket type for each data structure! ([], (), {})
  - That sets don’t keep order and only allow unique items
  - That tuple values can’t be changed after creation

Practice helps—you’ll get comfortable faster by trying simple problems!
